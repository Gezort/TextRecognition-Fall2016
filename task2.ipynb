```python
import re
regexp = re.compile('(\s)|([!,.\-\\?/\"\'])', re.UNICODE)
emails['words'] = emails.RawText.apply(lambda x : re.sub(regexp, ' ', x).split(' '))
```

```python
data = emails.RawText.values
print(data[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=10, max_df=500, stop_words='english', ngram_range=(1,3), preprocessor=(
        lambda x : x.replace('\n', ' ')))
features = vectorizer.fit_transform(data)
```

```python
vocabulary = vectorizer.vocabulary_
list(vocabulary.items())[:10]
```

```python
bigrams = list(filter(lambda x : len(x.split()) == 2, vocabulary.keys()))
bigrams = [(x, vocabulary[x]) for x in bigrams]
bigrams.sort(key=(lambda x:x[1]), reverse=True)
```

```python
bigrams[:20]
```

```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
email_words = emails.words.values
```

```python
finder = BigramCollocationFinder.from_documents(email_words)
finder.nbest(bigram_measures.pmi, 10)
```

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
clust = KMeans(n_clusters=50)
```

```python
clusters = clust.fit_predict(features)
```

Возьмем центры кластеров. Отсортируем n-граммы по встречаемости в этом кластере. Выведем топ сколько-то

```python
all_words = np.array(list(vocabulary.keys()))
for cl_cent in clust.cluster_centers_:
    cl_words = all_words[cl_cent > 0]
    cl_cent = (cl_cent > 0)
    indexes = np.argsort(cl_cent)
    n = int(len(cl_cent) * 0.001)
    cl_words = all_words[indexes][-n : ]
    print (cl_words)
```

```python
from gensim import matutils
from gensim.models.ldamodel import LdaModel
from gensim.corpora.dictionary import Dictionary
import pyLDAvis.gensim
```

```python
vec = CountVectorizer(min_df=10, max_df=500, stop_words='english')
X = vec.fit_transform(emails.RawText)
vocab = vec.get_feature_names()
vocab_dict = dict([(i, s) for i, s in enumerate(vocab)])
corpus = matutils.Sparse2Corpus(X)
```

```python
%time lda = LdaModel(corpus, num_topics=10, passes=30, id2word=vocab_dict)
```

```python
dictionary = Dictionary.from_corpus(corpus, vocab_dict)
corpus = matutils.Sparse2Corpus(corpus.sparse.T)
```

```python
pyLDAvis.enable_notebook()
vis_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```
